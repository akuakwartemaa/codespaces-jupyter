**Objective**: The primary aim of this project is to evaluate the safety of the treatment compared to the control group and visualize the overall profile of adverse events across different population groups. 1) **Hypothesis**:

**Null Hypothesis (H0)**: There is no difference in the incidence of adverse events between the treatment and control groups.

**Alternative Hypothesis (H1**): The incidence of adverse events is higher in the treatment group compared to the control group.

2) **Public Health Significance**:Assessing the safety of a new treatment compared to standard care is crucial for patient care and regulatory approval.

Understanding the adverse events profile helps healthcare providers and regulatory agencies make informed decisions about the risks and benefits of the treatment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the dataset
adae_data = pd.read_csv('/content/ADAE.csv')

 #Get the number of rows and columns
num_rows, num_cols = adae_data.shape

# Print the number of rows
print("Number of rows in the dataset:", num_rows)


# Group by treatment arm and calculate the count of adverse events for each arm
arm_counts = adae_data.groupby('ARM')['USUBJID'].count().reset_index()
print (arm_counts)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a DataFrame with the number of adverse cases in each treatment arm
data = {
    'ARM': ['Active control', 'Placebo', 'Test drug 140mg', 'Test drug 210mg'],
    'Adverse Cases': [427, 243, 636, 784]
}
df = pd.DataFrame(data)

# Perform chi-square test
observed_values = df['Adverse Cases'].values
expected_values = [sum(observed_values) / len(observed_values)] * len(observed_values)
chi2, p_value, _, _ = chi2_contingency([observed_values, expected_values])

# Set the significance level
alpha = 0.05

# Print the results
print(f'Chi-square statistic: {chi2}')
print(f'P-value: {p_value}')

# Check if the p-value is less than the significance level
if p_value < alpha:
    print('Reject the null hypothesis. There is a significant difference in adverse cases among treatment arms.')
else:
    print('Fail to reject the null hypothesis. No significant difference in adverse cases among treatment arms.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already have arm_counts DataFrame from the code you provided

# Plotting the results
plt.figure(figsize=(8, 6))
plt.bar(arm_counts['ARM'], arm_counts['USUBJID'], color='skyblue')
plt.xlabel('Treatment Arm')
plt.ylabel('Number of Adverse Events')
plt.title('Adverse Events Count by Treatment Arm')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```


3)**RESULTS/FINDING**:Based on the provided data, Out of the 2090 cases, we can observe that the test drug 210mg has the highest number of adverse cases (784), followed by the test drug 140mg (636), active control (427), and placebo (243).

After analyzing the ADAE dataset, we found that the treatment group had a higher incidence of adverse events compared to the control group. **Test Drug 210mg** has the highest number of adverse events (784), followed by **Test Drug 140mg** (636), **Active Control** (427), and **Placebo** (243).
The difference in adverse event incidence between the treatment and control groups was statistically significant (p < 0.05).
